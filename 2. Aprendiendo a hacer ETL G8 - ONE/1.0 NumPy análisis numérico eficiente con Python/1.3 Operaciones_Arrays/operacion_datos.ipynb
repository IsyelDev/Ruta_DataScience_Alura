$$\alpha = pendiente$$
$$\beta = interseccion$$
$$n = observaciones$$
$$y = precio$$
$$x = fecha$$
La solución para la pendiente es:
$$\alpha = \frac{n \sum xy - \sum x \sum y}{n \sum x^2 - 
(\sum x)^2} $$

Y para la intersección:
$$\beta = \frac{\sum y - \alpha \sum x}{n} $$
Y obtenemos la ecuación de la recta:
$$y = \alpha x + \beta $$

```python
n = np.size(ciudades["Moscu"])
y = ciudades["Moscu"]
x = fechas
print(fechas)
print(y)
```

```python
a = (n * np.sum(x * y) - np.sum(x) * np.sum(y)) / (n * np.sum(x**2) - np.sum(x)**2)
b = np.mean(y)- a*np.mean(x)
print(a)
print(b)
```

```python
y = a * x + b
y
```
