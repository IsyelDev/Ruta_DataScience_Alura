```python
import pandas as pd
import numpy as np
```

```python
datos_manzana = np.loadtxt("..//data/1.1 manzanas.csv",delimiter=",",usecols=np.arange(1,88,1))
datos_manzana
```

```python
datos_transpuestos= datos_manzana.T 
datos_transpuestos
```

```python
fechas = np.arange(1,88,1)
fechas
```

```python
precios = datos_transpuestos[:,1:]
```

```python
Moscu = precios[:,0]
Kaliningrado = precios[:,1]
Petesburgo = precios[:,2]
Krasnodar = precios[:,3]
Ekaterimburgo	 = precios[:,4]
```

```python
import matplotlib.pyplot as plt

ciudades = {
    "Moscu": Moscu,
    "Kaliningrado": Kaliningrado,
    "Petesburgo": Petesburgo,
    "Krasnodar":Krasnodar,
    "Ekaterimburgo":Ekaterimburgo	
}

for nombre,datos in ciudades.items():
    plt.figure()
    plt.plot(np.arange(1,len(datos)+1,1),datos)
    plt.title(f"Temperatura en {nombre}")
    plt.xlabel("Días")
    plt.ylabel("Temperatura")
    plt.grid(True)
    plt.show()
```

```python
ciudades["Moscu"].shape
```

```python
subconjuntos = {
    "Moscu_2013": ciudades["Moscu"][0:12],
    "Moscu_2014": ciudades["Moscu"][12:24],
    "Moscu_2015": ciudades["Moscu"][24:36],
    "Moscu_2016": ciudades["Moscu"][36:48],
    "Moscu_2017": ciudades["Moscu"][48:60],
    "Moscu_2018": ciudades["Moscu"][60:72],
    # etc.
}
subconjuntos
```

```python
for nombre , valores in subconjuntos.items():
    plt.plot(np.arange(1,len(valores)+1),valores)
    plt.legend(["2013","2014","2015","2016","2017","2018"])
```

```python
np.array_equal(subconjuntos["Moscu_2013"],subconjuntos["Moscu_2014"])
np.allclose(subconjuntos["Moscu_2013"],subconjuntos["Moscu_2014"],3)
```
