```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
url2 = "https://gist.githubusercontent.com/delapazcampos/2112426072daad307324bd1f0e6278b6/raw/3b071216795725878ca5e403f56171c3117ef546/naranja.csv"
tabla_naranjas = np.loadtxt(url2,delimiter=",",skiprows=1,usecols=np.arange(1,6,1))
tabla_naranjas
```

```python
diametro_naranja = np.array(tabla_naranjas[:5000,0])
diametro_toronjas = np.array(tabla_naranjas[5000:,0])
print(diametro_naranja)
print(diametro_toronjas)
```

```python
peso_naranja = np.array(tabla_naranjas[:5000,1])
peso_toronjas = np.array(tabla_naranjas[5000:,1])
print(peso_naranja)
print(peso_toronjas)
```

```python
plt.plot(diametro_toronjas,peso_toronjas)
plt.plot(diametro_naranja,peso_naranja)
```
