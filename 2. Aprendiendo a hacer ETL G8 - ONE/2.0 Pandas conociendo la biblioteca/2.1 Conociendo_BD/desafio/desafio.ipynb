```python
import pandas as pd

url =r"../../data/colonias.csv"
# Visualiza las primeras 7 filas del DataFrame y las últimas 5.

colonias = pd.read_csv(url,sep=";")
colonias
```

```python
colonias.head()
```

```python
valor_promedio = colonias.groupby("Tipo")["Valor"].mean().reset_index()
valor_promedio.rename(columns={"Valor": "promedio"}, inplace=True)
valor_promedio

```

```python
promedio_colonias = colonias.groupby("Tipo")["Valor"].mean().reset_index()
promedio_colonias.rename(columns={"Valor":"promedio"},inplace=True)
promedio_colonias
```

```python
promedio_colonia = colonias.groupby("Colonia")["Valor"].mean().reset_index()
promedio_colonia.rename(columns={"Valor":"Promedio"}, inplace=True)
promedio_colonia["Promedio"] = promedio_colonia["Promedio"].round(1)
promedio_colonia
```

```python
ambas_columnas = colonias.groupby(["Tipo","Colonia"])["Valor"].mean().reset_index()
ambas_columnas.rename(columns={"Valor":"PROMEDIO"}, inplace=True)
ambas_columnas
```

```python
propiedades = colonias.groupby(["Tipo"])["Valor"].count().reset_index()
propiedades.rename(columns={"Valor":"contador"},inplace=True)
propiedades
```

```python
colonias_filtrada = colonias[colonias["Area"]>2450]
mayor = colonias_filtrada.groupby("Tipo")["Valor"].mean().reset_index()
mayor
```
