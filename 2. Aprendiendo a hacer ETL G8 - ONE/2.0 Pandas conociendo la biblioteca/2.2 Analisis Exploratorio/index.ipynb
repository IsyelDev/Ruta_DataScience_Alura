```python
import pandas as pd

url =r"../../../2. Aprendiendo a hacer ETL G8 - ONE/2.0 Pandas conociendo la biblioteca/data/colonias.csv"
# Visualiza las primeras 7 filas del DataFrame y las últimas 5.

colonias = pd.read_csv(url,sep=";")
colonias
```

```python
colonias["Valor"].mean()
```

```python
colonias.groupby('Tipo').mean(numeric_only=True).mean()
```

```python
colonias.groupby('Tipo')['Valor'].mean()
```

```python
import matplotlib.pyplot as plt
df_tipoo = colonias.groupby("Tipo")[["Valor"]].mean().sort_values("Valor")
df_tipoo.plot(kind="barh" , figsize=(12,8) ,color="purple")
```
