```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
datos = pd.read_csv("../data/inmigrantes_canada.csv")
datos.head(5)
```

```python
datos.shape
```

```python
datos.info()
```

```python
datos.set_index('Pais',inplace=True)
datos
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
colombia = datos.loc["Colombia",anos]
colombia
```

```python
col_dict = {
    "Anos":colombia.index.tolist(),
    "Inmigrantes":colombia.values.tolist(),
            }
col_dict
```

```python
datos_col = pd.DataFrame(col_dict)
datos_col.head()
```

```python
plt.figure(figsize=(10,4))
plt.plot(datos_col["Anos"] , datos_col["Inmigrantes"])
plt.title("nuevo")
plt.xlabel("anos")
plt.ylabel("inmigrantes")
plt.grid(True)
plt.xticks(["1980","1985","1990","1995","2000","2005"])
plt.show()
```
