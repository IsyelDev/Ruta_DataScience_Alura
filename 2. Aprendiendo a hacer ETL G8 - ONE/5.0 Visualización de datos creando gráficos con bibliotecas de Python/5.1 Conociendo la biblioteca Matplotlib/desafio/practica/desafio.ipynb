```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
datos = pd.read_csv("../../../data/inmigrantes_canada.csv")
datos.head(3)
```

```python
datos.set_index("Pais" , inplace=True)
```

```python
anos = list(map(str,range(1980,2014)))
anos
```

```python
comparacion = datos.loc[['Brasil', 'Argentina'],anos]
comparacion
```

```python
comparacion = comparacion.T
```

```python
plt.plot(comparacion["Brasil"],label="Brasil",marker="o")
plt.plot(comparacion["Argentina"],label="Argentina",marker="o")
plt.xticks(rotation=45)
plt.xticks(ticks=[str(year) for year in range(1980, 2014, 5)])  # cada 5 años
plt.tight_layout()
plt.legend()
plt.title('Inmigración de Brasil y Argentina a Canadá')
plt.xlabel('Año')
plt.ylabel('Número de Inmigrantes')
plt.grid(True)
plt.show()
```
