```python
import matplotlib.pyplot as plt

# Coordenadas de los puntos
x = [2, 1, 0, -1, -2]
y = [5, 2, -1, -4, -7]

# Crear gr√°fico
plt.scatter(x, y)
plt.axhline(0, color='blue')  # Eje X
plt.axvline(0, color='red')  # Eje Y
plt.grid(True)
plt.title('Puntos en el plano cartesiano')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```
